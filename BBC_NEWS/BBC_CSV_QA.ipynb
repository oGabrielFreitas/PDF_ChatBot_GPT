### Importando mais ferramentas

```python
from langchain import OpenAI, ConversationBufferMemory, VectorstoreSearch
from langchain.agents import initialize_agent, Tool
from langchain.agents import Agent

# RetrievalQA tool 
retrieval_qa = Tool(name="RetrievalQA", 
                    func=VectorstoreSearch(vectorstore=vectordb).run_search)

# Conversation tool
memory = ConversationBufferMemory() 
conversation = Tool(name="Conversation",
                    func=memory.get_memory) 

# WebGPT tool                   
web_gpt = Tool(name="WebGPT",
               func=OpenAI(temperature=0).call)
               
# Create list of tools              
tools = [retrieval_qa, conversation, web_gpt]

# Initialize agent with tools
agent = initialize_agent(tools, 
                         agent=Agent.REACT_TOOL_DESCRIPTION,
                         memory=memory) 

# Use the agent
agent.run("What is the capital of Brazil?")
```
